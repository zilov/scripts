```python
import os
import os.path
import glob 

def mkdir(path):
    if not os.path.exists(path):
        os.mkdir(path)
    
def mkfile(path):
    if not os.path.exists(path):
        os.system(f"touch {path}")
```

```python
cluster_file = "/media/eternus1/nfs/projects/sigma/clusters/cluster_1_NZ_UAWN01000005_292699_293367_CDS_46660.fasta"
cluster_prefix = cluster_file.split("/")[-1].split(".")[0]
workdir = "/media/eternus1/nfs/projects/sigma/clusters/"
mkdir(workdir)
```

```python
cluster_list = glob.glob("/media/eternus1/nfs/projects/sigma/clusters/*.fasta")
cluster_list.sort()
```

```python
# muscle 
for cluster_file in cluster_list:
    cluster_prefix = cluster_file.split("/")[-1].split(".")[0]
    muscle_input = cluster_file
    muscle_dir = workdir + "muscle/"
    mkdir(muscle_dir)
    muscle_output = muscle_dir + cluster_prefix + '.msa'
    mkfile(muscle_output)
    #print(muscle_output)
    #muscle -in file -out file_msa -maxiters 2
    muscle_command = f"muscle -in {muscle_input} -out {muscle_output} -maxiters 2"
    print(muscle_command)
    os.system(muscle_command)
```

```python
#hmmbuild file.hmm file_msa
hmm_in_list = glob.glob("/media/eternus1/nfs/projects/sigma/clusters/muscle/*.msa")
hmm_in_list.sort()
for hmm_input in hmm_in_list:
    cluster_prefix = hmm_input.split("/")[-1].split(".")[0]
    hmm_dir = workdir + "hmm/"
    mkdir(hmm_dir)
    hmm_output = hmm_dir + cluster_prefix + ".hmmbuild"

    hmm_command = f"hmmbuild {hmm_output} {hmm_input}"
    print(hmm_command)
    os.system(hmm_command)
```
